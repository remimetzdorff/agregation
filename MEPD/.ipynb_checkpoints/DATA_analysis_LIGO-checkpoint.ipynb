```python
folder = "figures/data/"
filename = "H-H1_LOSC_4_V2-1126259446-32.txt"
sample_rate = 2**12
duration = 32
t = np.linspace(0, duration, duration*sample_rate)
h = np.loadtxt(folder+filename)

plt.plot(t, h)
```

```python
def make_fft(x,y):
    N = len(x)
    T = x[1]-x[0]
    yf = fft(y)
    xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
    return xf, 2.0/N * np.abs(yf[0:N//2])

xf, yf = make_fft(t, h)
plt.loglog(xf, yf)
plt.xlim(10,2000)
```
