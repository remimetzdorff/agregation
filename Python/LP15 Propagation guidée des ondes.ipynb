```python
def beta(omega, a, n=2):
    return np.sqrt((omega/cst.c)**2 - (n*np.pi/a)**2)

omega = np.linspace(0,500e9,10001)
a = 1e-2
omega_c = np.pi*cst.c/a
```

```python
plt.figure(figsize=(6,4))

plt.plot(omega/omega_c, cst.c*beta(omega, a, n=0)/omega_c, "--")

for n in [1,2,3,4]:
    plt.plot(omega/omega_c, cst.c*beta(omega, a, n=n)/omega_c, label="n="+str(int(n)))
plt.legend(loc="upper left")
plt.xlabel("$\omega/\omega_c$")
plt.ylabel(u"$c\\beta/\omega_c$")

mysavefig("dispersion.png")
```
