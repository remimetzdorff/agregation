```python
theta = np.linspace(0, 2*np.pi, 1001)

def conic(theta, p=1, e=0):
    return p/(1+e*np.cos(theta))

r = conic(theta, p=1, e=0.1)
x = r*np.cos(theta)
y = r*np.sin(theta)

plt.figure()
sps = (1,1)
ax = plt.subplot2grid(sps, (0,0))

ax.plot(x, y)
ax.set_aspect("equal")
```
