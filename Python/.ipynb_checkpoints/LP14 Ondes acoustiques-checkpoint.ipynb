```python
%pylab qt
%matplotlib inline

import pandas
import fitting

def celerite(temp):
    return sqrt(1.4*8.314*temp/28.8e-3)

T0 = 273.16
```

## Vitesse du son dans l'air

$M = 28,8\ g/mol$, $\gamma = 1,4$

```python
temp = np.linspace(250, 310, 1001)

plt.plot(temp-T0, celerite(temp))

T_amb   = T0 + 25
uT_amb  = 1

freq        = 40e3
wavelength  = 8.7e-3
uwavelength = 0.1e-3

c_theo  = celerite(T_amb)
uc_theo = c_theo*uT_amb/T_amb/2
c_exp   = wavelength*freq
uc_exp  = c_exp*uwavelength/wavelength

plt.errorbar([T_amb-T0], [c_exp], xerr=[uT_0], yerr=[uc_exp], fmt="o")

print("Vitesse THEORIQUE du son dans l'air     = %5.1f +/- %3.1f m/s"%(c_theo, uc_theo))
print("Vitesse EXPERIMENTALE du son dans l'air = %5.0f +/- %3.0f m/s"%(c_exp, uc_exp))

plt.xlabel("Température (°C)")
plt.ylabel("c (m/s)")
```

```python
plt.figure(figsize=(6,3))
sps = (2,3)

x = np.linspace(0,1, 1001)

for n in [1,2,3]:
    ax1 = plt.subplot2grid(sps, (0,n-1))
    ax2 = plt.subplot2grid(sps, (1,n-1))
    ax1.set_title("n=%.0g"%(n))
    ax1.set_xticklabels([])
    if n!=1:
        ax1.set_yticklabels([])
        ax2.set_yticklabels([])
    if n==1:
        ax1.set_ylabel("surpression")
        ax2.set_ylabel("vitesse")
    ax2.set_xlabel("x")
    for a in [-1,-0.5,0.5,1]:
        p = np.sin(n*np.pi*x)
        ax1.plot(x, a*p, "C0")
        v = np.cos(n*np.pi*x)
        ax2.plot(x, a*v, "C1")
plt.tight_layout()
plt.savefig("sound_tube_open.png", bbox_inches="tight", dpi=200)
```

```python
plt.figure(figsize=(4,3))
sps = (2,2)

x = np.linspace(0,1, 1001)

for n in [1,2]:
    ax1 = plt.subplot2grid(sps, (0,n-1))
    ax2 = plt.subplot2grid(sps, (1,n-1))
    if n==1:
        ax1.set_title("ouvert-fermé")
    if n==2:
        ax1.set_title("ouvert-ouvert")
    ax1.set_xticklabels([])
    if n!=1:
        ax1.set_yticklabels([])
        ax2.set_yticklabels([])
    if n==1:
        ax1.set_ylabel("surpression")
        ax2.set_ylabel("vitesse")
    ax2.set_xlabel("x")
    for a in [-1,-0.5,0.5,1]:
        p = np.sin(n/2*np.pi*x)
        ax1.plot(x, a*p, "C0")
        v = np.cos(n/2*np.pi*x)
        ax2.plot(x, a*v, "C1")
plt.tight_layout()
plt.savefig("sound_tube_open_close.png", bbox_inches="tight", dpi=200)
```
