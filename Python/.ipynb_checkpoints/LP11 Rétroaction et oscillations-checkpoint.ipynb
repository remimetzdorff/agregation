```python
%pylab qt
%matplotlib inline

import scipy.constants as cst

def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)

def radius(theta, p=1, e=1, theta0=0):
    return p / (1+e*np.cos(theta-theta0))

def angular_speed(theta, p=1, e=1, theta0=0, C=1):
    return C/radius(theta, p=p, e=e, theta0=theta0)**2

def coordinates(theta, p=1, e=1, theta0=0):
    r = radius(theta, p=p, e=e, theta0=theta0)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return (x, y)

def fill_theta(dt, p=1, e=1, theta0=0, C=1, start=0):
    theta = [start]
    while theta[-1]<2*np.pi:
        if len(theta) > 1000: break
        theta.append(theta[-1] + angular_speed(theta[-1], p=p, e=e, theta0=theta0, C=1) * dt)
    return np.array(theta[:-1])
```

```python
T = 92.69 * 60
Mt = 5.972*1e24
Rt = 6371e3
r = (cst.G*Mt*T**2/4/np.pi**2)**(1/3)
h = r-Rt
print(h*1e-3)
```

```python
plt.figure(figsize=(8,8))
sps = (1,1)
ax = plt.subplot2grid(sps, (0,0))
ax.set_aspect("equal")

plt.plot([0,0], [0,0], "ok", markersize=20)

p, e, theta0, C = 1, 0.5, 0, 1

theta = np.linspace(0,2*np.pi, 1001)
x, y  = coordinates(theta, p=p, e=e, theta0=theta0)
plt.plot(x,y)

theta = fill_theta(0.2525, p=p, e=e, theta0=theta0, C=C)
x, y  = coordinates(theta, p=p, e=e, theta0=theta0)
plt.plot(x,y,"o")
print(len(theta))
```

```python
plt.figure(figsize=(15,3))
sps = (1,1)
ax1 = plt.subplot2grid(sps, (0,0))
#ax2 = plt.subplot2grid(sps, (0,1))
#ax3 = plt.subplot2grid(sps, (0,2))
ax1.set_aspect("equal")
#ax2.set_aspect("equal")
#ax3.set_aspect("equal")    

p, C, theta = 1, 1, 0
dt = 0.25

for e, ax in zip([0.8], [ax1]):
    ax.plot([0,0], [0,0], "oC1", markersize=10)
    ax.set_xticks([])
    ax.set_yticks([])
    
    thetas = np.linspace(0,2*np.pi, 1001)
    x, y  = coordinates(thetas, p=p, e=e, theta0=theta0)
    ax.plot(x,y)
    
    for i, start in enumerate([0, np.pi]):
        theta = fill_theta(dt, p=p, e=e, theta0=theta0, C=C, start=start)
        theta1 = theta[1]
        angle = []
        for theta in thetas:
            if theta < theta1 and theta > start:
                angle.append(theta)
        newx, newy = coordinates(np.array(angle), p=p, e=e, theta0=theta0)
        newxx = list(newx)
        newyy = list(-newy)
        newx = list(newx)
        newy = list(newy)
        newx.append(0)
        newy.append(0)
        newxx.reverse()
        newyy.reverse()
        newx += newxx
        newy += newyy
        x1, y1  = coordinates(np.array([-theta1, theta1]), p=p, e=e, theta0=theta0)
        ax.plot(x1,y1, "oC"+str(i+2))
        ax.fill(newx, newy, "oC"+str(i+2), alpha=0.5)

mysavefig("areolar_speed_"+str(e)+".png")
```

```python
planete = ["Mercure", "Vnus"  , "Terre"    , "Mars"    , "Jupiter", "Saturne" , "Uranus"  , "Neptune" ]
a       = [57909230 , 108209500, 149597887.5,  227944000, 778340000, 1426700000, 2870700000, 4498400000]
T       = [87.95565 , 224.667  , 365.256363 , 686.885   , 4332.01  , 10754     , 30698     , 60216.8   ]

a = np.array(a)*1e3
T = np.array(T)*3600*24
Ms = 1.9891*1e30

x = np.logspace(10.5,13, 101)
y = np.sqrt(4*np.pi**2*x**3/cst.G/Ms)

plt.figure(figsize=(6,4))
ax = plt.subplot2grid((1,1), (0,0))

ax.loglog(a, T, "o")
ax.loglog(x,y)
ax.set_xlabel("Demi-grand axe (m)")
ax.set_ylabel("Priode (s)")

for name, aa, t in zip(planete, a, T):
    ax.annotate(name, (aa*1.2, t/1.5))
    
mysavefig("kepler_third_law.png")
```
