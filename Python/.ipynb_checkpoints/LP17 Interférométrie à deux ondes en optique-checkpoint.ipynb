# LP17 Intreférométrie à deux ondes en optique

```python
%pylab qt
%matplotlib inline

import fitting
from fitting import find_exponent
from fitting import display_readable
def display_result(name, param, uparam, unit):
    print(name + " = " + display_readable(param, uparam) + unit)

def interfrange(wavelength, a, d):
    """valeur théorique de l'interfrange"""
    return wavelength * d / a

def uncertainty(x, ux):
    uy2 = 0
    for val, uval in zip(x, ux):
        uy2 += (uval/val)**2
    return np.sqrt(uy2)

def contrast_from_fit(freq, scale, phi, offset):
    """ returns (Imax-Imin)/(Imax+Imin) for a given (scale, offset) set """
    Imax = offset + scale
    Imin = offset - scale
    return (Imax-Imin)/(Imax+Imin)
```

## Interférences constructives / Interfrérences destructives

```python
x = np.linspace(0, 5, 1001)

fig = plt.figure()

for i, phi in enumerate(np.linspace(0,2*np.pi-2*np.pi/25, 24)):
    fig.clear()
    y1 = np.sin(2*np.pi*x)
    y2 = np.sin(2*np.pi*x+phi)

    plt.plot(x, y1, alpha=0.5, label="onde 1")
    plt.plot(x, y2, alpha=0.5, label="onde 1")
    plt.plot(x, y1+y2, label="onde 1 + 2")

    plt.xlim(0,5)
    plt.ylim(-2.1,2.1)
    plt.xticks([])
    plt.yticks([])
    plt.legend(loc="upper right")
    
    #plt.savefig(str(i)+".png", dpi=200, bbox_inches="tight")
```

## Mesure de l'interfrange

```python
wavelength  = 532e-9        # longueur d'onde du laser en m
uwavelength =   1e-9        # incertitude sur cette valeur
a           = 100e-6        # distance entre les fentes en m
ua          =   1e-6        # incertitude sur cette valeur
d           =   2           # distance entre les fentes et l'écran en m
ud          =   1e-2        # incertitude sur cette valeur

# Valeur théorique de l'interfrange
i  = interfrange(wavelength, a, d)
ui = i*uncertainty([wavelength, a, d], [uwavelength, ua, ud])
display_result("interfrange théorique", i, ui, "m")

# Valeur expérimentale
i_mes  = 5e-3               # interfrange mesuré
ui_mes = 1e-3               # incertitude sur cette valeur
display_result("interfrange mesuré   ", i_mes, ui_mes, "m")
```

```python
### TEST ###
from random import gauss

freq   = 0.2148
phi    = np.pi/2
scale  = 2.20454
offset = scale

fitter = fitting.Fit("sin")

x = np.linspace(-5,5, 101)
y = fitter.func(x, freq, scale, phi, offset)
noisey = np.zeros(len(y))
for i,n in enumerate(noisey):
    noisey[i] = gauss(0,scale/10)
y += noisey
    
plt.plot(x,y)

fitter = fitting.Fit("sin", x=x, y=y, uy=noisey)
fit_params = fitter.fit(manualguess_params=None)

plt.plot(x, fitter.sin(x, *fit_params))

print(contrast_from_fit(*fit_params))
```
