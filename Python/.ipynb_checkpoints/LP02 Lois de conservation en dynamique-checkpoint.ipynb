```python
%pylab qt
%matplotlib inline

import scipy.constants as cst

def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)

def radius(theta, p=1, e=1, theta0=0):
    return p / (1 + e*np.cos(theta-theta0))

def coordinates(theta, p=1, e=1, theta0=0):
    r = radius(theta, p=p, e=e, theta0=theta0)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return (x, y)
```

```python
plt.figure(figsize=(6,4))
sps = (1,1)
ax = plt.subplot2grid(sps, (0,0))
ax.set_aspect("equal")
theta = np.linspace(0,2*np.pi, 1001)

m1, m2, e = 0.7, 0.3, 0.7

x, y = coordinates(theta, e=e)
x1, y1 = m1*x, m1*y
x2, y2 =  -m2*x, -m2*y
ax.plot(x1, y1, "C0")
ax.plot(x2, y2, "C1")
ax.plot(x, y, "--C2")

x, y = coordinates(3*np.pi/4, e=e)
x1, y1 = m1*x, m1*y
x2, y2 =  -m2*x, -m2*y
ax.plot([0,x1], [0,y1], "-oC0")
ax.annotate("$M_1$", (x1+0.1, y1+0.1), color="C0")
ax.plot([0,x2], [0,y2], "-oC1")
ax.annotate("$M_1$", (x2+0.1, y2-0.2), color="C1")
ax.plot([0,x], [0,y], "--oC2")
ax.annotate("$M$", (x-0.2, y-0.2), color="C2")

ax.plot([0], [0], "ok")
ax.annotate("$G$", (0.1,0))

ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(-4,1.5)
ax.set_ylim(-1.5, 1.5)

mysavefig("mobile_fictif_1.png")
```
