```python
%pylab qt
%matplotlib inline

import scipy.constants as cst

def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)
```

```python
##### Oil #####
d = 5e-10
e = 1/40 * cst.e

print(e/d**2)
```

```python
##### Water #####
d = 2e-10
e = 0.2 * cst.e

print(e/d**2)
```

```python
##### Mercury #####
d = 4e-10
e = 1 * cst.e

print(e/d**2)
```
