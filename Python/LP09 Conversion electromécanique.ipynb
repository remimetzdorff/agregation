```python
%pylab qt
%matplotlib inline

import scipy.constants as cst

def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)
    
def champ_mag_rot(f, t, theta, scale=0.2):
    return scale*np.cos(2*np.pi*f*t-theta)
def champ_mag(f, t, theta, scale=0.2):
    return scale*np.cos(2*np.pi*f*t)*np.cos(theta)
```

```python
fig = plt.figure(figsize=(8,8))

def plot_fieldmap1(f=1, t=0, fig=fig):
    sps = (1,1)
    ax = plt.subplot2grid(sps, (0,0))
    ax.set_aspect("equal")
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_xticklabels([])
    ax.set_yticklabels([])

    time = np.linspace(0,2*np.pi,1001)
    x = np.cos(time)
    y = np.sin(time)
    ax.plot(x, y, "--")

    n = 24
    thetas = np.linspace(0,2*np.pi-2*np.pi/n, n)
    xs = np.cos(thetas)
    ys = np.sin(thetas)
    plt.plot(xs, ys, "o")

    for theta in thetas:
        x = np.cos(theta)
        y = np.sin(theta)
        B = champ_mag(f,t,theta)
        Bx = B*np.cos(theta)
        By = B*np.sin(theta)
        plt.arrow(x, y, Bx, By, shape='full', length_includes_head=True, lw=2, head_width=.02, color="C2")
    B = champ_mag(f,t,0)
    plt.arrow(0, 0, 2*B, 0, shape='full', length_includes_head=True, lw=2, head_width=.02, color="C2")
    
fe = 25
T = 1
for i, t in enumerate(np.linspace(0,T-1/fe, T*fe)):
    plot_fieldmap1(t=t)
    mysavefig("statoric_field1_"+str(i)+".png")
```

```python
fig = plt.figure(figsize=(8,8))

def plot_fieldmap1(f=1, t=0, fig=fig):
    sps = (1,1)
    ax = plt.subplot2grid(sps, (0,0))
    ax.set_aspect("equal")
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_xticklabels([])
    ax.set_yticklabels([])

    time = np.linspace(0,2*np.pi,1001)
    x = np.cos(time)
    y = np.sin(time)
    ax.plot(x, y, "--")

    n = 24
    thetas = np.linspace(0,2*np.pi-2*np.pi/n, n)
    xs = np.cos(thetas)
    ys = np.sin(thetas)
    plt.plot(xs, ys, "o")

    for theta in thetas:
        x = np.cos(theta)
        y = np.sin(theta)
        B = champ_mag_rot(f,t,theta)
        Bx = B*np.cos(theta)
        By = B*np.sin(theta)
        plt.arrow(x, y, Bx, By, shape='full', length_includes_head=True, lw=2, head_width=.02, color="C0")

    B = champ_mag(f,t-1/4/f,0)
    plt.arrow(0, 0, 0, 2*B, shape='full', length_includes_head=True, lw=2, head_width=.02, color="C3", alpha=0.5)
    B = champ_mag(f,t,0)
    plt.arrow(0, 0, 2*B, 0, shape='full', length_includes_head=True, lw=2, head_width=.02, color="C2", alpha=0.5)
    plt.arrow(0, 0, 0.4*np.cos(2*np.pi*f*t), 0.4*np.sin(2*np.pi*f*t), shape='full', length_includes_head=True, lw=2, head_width=.02, color="C0")
    
for i, t in enumerate(np.linspace(0,T-1/fe, T*fe)):
    plot_fieldmap1(t=t)
    mysavefig("statoric_field_tot_"+str(i)+".png")
```
