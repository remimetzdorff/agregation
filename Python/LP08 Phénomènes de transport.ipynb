```python
%pylab qt
%matplotlib inline

import scipy.constants as cst
import random as rd
import fitting


def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)

def random_walk_2D(N, l=1):
    """2D random wal path generator, with N step of length l"""
    thetas = np.linspace(0,2*np.pi,101)[:-1]
    xs, ys = [0], [0]
    for i in range(N-1):
        theta = rd.choice(thetas)
        x = l*np.cos(theta)
        y = l*np.sin(theta)
        xs.append(x+xs[-1])
        ys.append(y+ys[-1])
    xs = np.array(xs)
    ys = np.array(ys)
    return xs, ys    
    
def mapper(x, y, d, N=51):
    """count the number of values with (x,y) coordinates around each point of a N**2 sized grid
       d value is the limit for the extremal bins -d,+d"""
    step = 2*d/(N-1)
    X = np.linspace(-d,d,N)
    Y = np.linspace(-d,d,N)
    Z = []
    for i in range(N):
        Z.append(np.zeros(N))
    Z = np.array(Z)
    for xval, yval in zip(y,x):
        xindex = int((N+1)/2 + xval//step)-1
        yindex = int((N+1)/2 + yval//step)-1
        if xindex < N-1 and xindex > 0:
            if yindex < N-1 and yindex > 0:
                Z[xindex, yindex] += 1
    return X,Y,Z

def probability(x, y, n, l=1):
    """2D gaussian propability to be at x,y position after a random walk with n step of length l"""
    N = len(x)
    Z = []
    for i in range(N):
        Z.append(np.zeros(N))
    Z = np.array(Z)
    for i, xval in enumerate(x):
        for j, yval in enumerate(y):
            r = np.sqrt(xval**2+yval**2)
            Z[i,j] = 2/n/l**2 * np.exp(-r**2/n/l**2)
    return Z
```

```python
plt.figure(figsize=(3,3))
ax = plt.subplot2grid((1,1), (0,0))

N = 2**6
xs, ys = random_walk_2D(N)
for n in [1,2,3,4,5,6]:
    ax.clear()
    ax.plot(xs[:2**n], ys[:2**n], '-k', alpha=0.5)
    ax.plot(xs[2**n-1], ys[2**n-1], 'ok')
    d = np.sqrt(N)
    ax.set_xlim(-d, d)
    ax.set_ylim(-d, d)
    ax.set_aspect("equal")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_xticks([])
    ax.set_yticks([])
    ax.annotate("step = %3.0f"%(2**n-1), (-d+0.05*d, d-0.12*d))
    mysavefig("random_walk_"+str(n)+".png")
```

```python
def rmax(N, l=1, display=True):
    xs, ys = random_walk_2D(N, l=l)
    r = np.sqrt(xs**2+ys**2)
    r = np.max(r)
    if display:
        plt.figure(figsize=(6,6))
        ax = plt.subplot2grid((1,1), (0,0))
        ax.plot(xs, ys, '-')
        ax.set_aspect("equal")
        thetas = np.linspace(0, 2*np.pi, 1001)
        ax.plot(r*np.cos(thetas), r*np.sin(thetas))
    return r

def rend(N, l=1, display=True):
    xs, ys = random_walk_2D(N, l=l)
    r = np.sqrt(xs**2+ys**2)
    r = r[-1]
    if display:
        plt.figure(figsize=(6,6))
        ax = plt.subplot2grid((1,1), (0,0))
        ax.plot(xs, ys, '-')
        ax.set_aspect("equal")
        thetas = np.linspace(0, 2*np.pi, 1001)
        ax.plot(r*np.cos(thetas), r*np.sin(thetas))
    return r

#rmax(1000)
rend(1000)
```

```python
def diffusion(Npart=2**12, Ntime=2**6, d=None, display=True):
    """simulate the diffusion of Npart particles for Ntime steps
       d is the limit for the mapping"""
    
    if d is None:
        d = 3*np.sqrt(Ntime)
    ncolor = 101
            
    X, Y = [], []
    for i in range(Npart):
        xs, ys = random_walk_2D(Ntime)
        xend, yend = xs[-1], ys[-1]
        X.append(xend)
        Y.append(yend)
    x, y, z = mapper(X,Y,d)
    ztheo = probability(x,y, Ntime)

    if display:
        plt.figure(figsize=(10,4))
        sps = (1,3)
        ax1 = plt.subplot2grid(sps, (0,0))
        ax2 = plt.subplot2grid(sps, (0,1))
        ax3 = plt.subplot2grid(sps, (0,2))
        for ax in [ax1, ax2, ax3]:
            ax.set_aspect("equal")
            ax.set_xlim(-d, d)
            ax.set_ylim(-d, d)
            ax.set_xticks([])
            ax.set_yticks([])
            ax.set_xlabel("x")
            ax.set_ylabel("y")
        ax1.set_title("Position\n(simulation)")
        ax2.set_title("Concentration\n(simulation)")
        ax3.set_title("Concentration\n(gaussienne)")
        ax1.plot(X, Y, '.k', alpha=0.05)
        ax1.annotate("steps = %4.0f"%(Ntime), (-d+0.05*d,d-0.11*d))
        ax2.contourf(x,y,z, ncolor)
        ax3.contourf(x,y,ztheo, ncolor)
        plt.tight_layout()
    
    return x,y,z

Ntimes = [116]
Ntimes = np.linspace(4, 2**9, 25)
for Ntime in Ntimes:
    Ntime=int(Ntime)
    print(Ntime)
    x,y,z = diffusion(Ntime=Ntime, d=25)
    mysavefig("diffusion2d_"+str(Ntime)+".png")
```

```python
plt.figure(figsize=(8,4))
sps = (1,2)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (0,1))

bw     = [] 
Ntimes = 2**np.linspace(2, 9, 15)
for Ntime in Ntimes:
    Ntime=int(Ntime)
    x,y,z = diffusion(Ntime=Ntime, display=False)

    position = x
    values   = []
    uvalues  = []
    for row in z:
        values.append(mean(row))
        uvalues.append(std(row))
    values  = np.array(values)
    uvalues = np.array(uvalues)
    
    fitter = fitting.Fit("gaussian", x=position, y=values)#uy=uvalues)
    fit_params, ufit_params = fitter.fit(manualguess_params=[0, np.sqrt(Ntime), np.max(values), 0], verbosemode=False)
    
    if Ntime in [2**2, 2**4, 2**6, 2**8, 2**10]:
        i = int(np.log2(Ntime)/2-1)
        ax1.plot(position, values/fit_params[2], "oC"+str(i), alpha=0.5)
        ax1.plot(position, fitter.gaussian(position, *fit_params)/fit_params[2], "C"+str(i), label=str(Ntime))
        ax1.legend(loc="upper left")
    bw.append(fit_params[1])
ax2.loglog(Ntimes, bw, 'o', label="simulations")
ax2.loglog(Ntimes, np.sqrt(Ntimes), label="$\propto\sqrt{t}$")

ax1.set_xlabel("x")
ax1.set_ylabel("Concentration (u.a.)")

ax2.set_xlabel("Temps (u.a.)")
ax2.set_ylabel("Distance (u.a.)")
ax2.legend(loc="upper left")

mysavefig("simulation_distance_diffusion.png")
```

```python
dt = 0.0005
dy = 0.0005
k = 10**(-4)
y_max = 0.04
t_max = 10
T0 = 100

def FTCS(dt,dy,t_max,y_max,k,T0):
    s = k*dt/dy**2
    y = np.arange(0,y_max+dy,dy) 
    t = np.arange(0,t_max+dt,dt)
    r = len(t)
    c = len(y)
    T = np.zeros([r,c])
    T[:,0] = T0
    for n in range(0,r-1):
        for j in range(1,c-1):
            T[n+1,j] = T[n,j] + s*(T[n,j-1] - 2*T[n,j] + T[n,j+1]) 
    return y,T,r,s

y,T,r,s = FTCS(dt,dy,t_max,y_max,k,T0)

for i in [2,20,200,2000,20000]:
    plt.plot(y, T[i])
```
