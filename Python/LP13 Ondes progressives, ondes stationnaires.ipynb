```python
%pylab qt
%matplotlib inline

import scipy.constants as cst

def mysavefig(name):
    plt.tight_layout()
    plt.savefig(name, bbox_inches="tight", dpi=200)
```

```python
plt.figure(figsize=(8,6))
sps = (2,1)
ax1 = plt.subplot2grid(sps, (0,0))
ax2 = plt.subplot2grid(sps, (1,0))

omega = 2*np.pi
k = 1.5*2*np.pi

x = np.linspace(0,1, 1001)
time = np.array(range(0,51))/51+1/50

for i, t in enumerate(time):
    ax1.clear()
    ax2.clear()

    rope = np.sin(k*x)*np.cos(omega*t)
    ec   = (np.sin(k*x)*np.sin(omega*t))**2
    ep   = (np.cos(k*x)*np.cos(omega*t))**2
    ax1.plot(x, rope, "C0")
    ax1.plot([0,1], [0,0], "ok")
    ax2.plot(x, ec, "C1", label="$E_c$")
    ax2.plot(x, ep, "C2", label="$E_p$")
    ax2.plot(x, ep+ec, "C3", alpha=0.25, label="$E_c+E_p$")
    
    ax1.set_ylim(-1.1,1.1)
    ax2.set_ylim(-0.1,1.1)
    for ax in [ax1, ax2]:
        ax.set_xlim(-0.1,1.1)
        
    ax1.set_xticklabels([])
    ax1.set_ylabel("y (u.a.)")
    ax2.set_xlabel("x (u.a.)")
    ax2.set_ylabel("Energie (u.a.)")
    ax2.legend(loc="upper right")
    
    mysavefig("rope_"+str(i)+".png")
```
